```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def cal_weight(X, Y):
    W = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
    Yhat = np.dot(X, W)
    print("Predictions: ", Yhat)
    
    d1 = Y - Yhat
    d2 = Y - Y.mean()
    r2 = 1 - d1.dot(d2) / d2.dot(d2)
    print("r-squared: ", r2)
    return Yhat
```
